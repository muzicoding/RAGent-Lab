# 2. 提示词工程
## 2.1 提示词工程概念与技巧

**提示工程 (Prompt Engineering)** 是指在不更新模型权重的情况下，通过设计与大模型交互的指令（Prompt）来引导其行为，以获得所需结果。

**核心思想**: 清晰地表达你的需求。

**技巧**
1. 详细的描述:
    - 差: “写一封情书”
    - 好: “用温柔的话语写一封不少于500字的情书，表达我的仰慕和思念。”
2. 让模型充当某个角色:
    - “我需要你充当一个AI算法面试官的角色，对我进行面试，一次问一个问题。”
3. 使用分隔符标明输入的不同部分:
    - “用20个字符总结由三引号分割的文本。"""在此插入文本"""”
4. 对任务指定步骤:
    - “步骤1: 用一句话总结文本，并加上前缀 'Summary'。
    - 步骤2: 将摘要翻译成英语，并添加前缀 '翻译：'。”
5. 提供例子 (Few-shot Learning):
    - 先给出输入和期望输出的示例，再让模型处理新输入。
6. 使用参考文本作答 (降低幻觉):
    - “根据下文中三重引号引起来的文章来回答问题。如果找不到答案，请写 ‘我找不到答案’。”

> **总结:** 提示词优化是所有大模型应用开发的基础。一个好的提示词，甚至能让基础模型的输出媲美微调后的模型。
